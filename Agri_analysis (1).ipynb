```python
import seaborn as sns
corr_matrix = df2.corr()

plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()
```

```python
correlation_threshold = 0.1

# Identifying low correlated features
low_correlation_features_content = corr_matrix[abs(corr_matrix['Modal Price (Rs./Quintal)']) < correlation_threshold].index

low_correlation_features_content
```

```python
df2 = df2.drop(low_correlation_features_content, axis=1)
```

```python
df2.head(2)
```

The data is split into training and testing sets, with the data from 2019 to 2020 used for testing and the rest used for training purpose.

```python
train_data = df2[df2['Year'] < 2019]
test_data = df2[(df2['Year'] >= 2019) & (df2['Year'] <= 2020)]
```

```python
x_train = train_data.drop('Modal Price (Rs./Quintal)', axis=1)
x_test = test_data.drop('Modal Price (Rs./Quintal)', axis=1)
y_train = train_data['Modal Price (Rs./Quintal)']
y_test = test_data['Modal Price (Rs./Quintal)']
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
y_test.shape
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
```

```python
x_train_scaled = scaler.fit_transform(x_train)
```

```python
x_test_scaled = scaler.transform(x_test)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
```

```python
models = [
    ("Linear Regression", LinearRegression()),
    ("Decision Tree", DecisionTreeRegressor()),
    ("Random Forest", RandomForestRegressor())
]

# Training and evaluation of models
for name, model in models:
    model.fit(x_train_scaled, y_train)
    y_pred = model.predict(x_test_scaled)
    R2 = r2_score(y_test, y_pred)
    print(f"{name} R2 score: {R2}")
```

From the above, it can be conclude that the Linear regression model is of high accuracy.

```python
from sklearn.model_selection import cross_val_score, KFold

model = LinearRegression()

# Define the number of folds for cross-validation
k_folds = KFold(n_splits=5, shuffle=True, random_state=42)

# Cross-validation
cv_scores = cross_val_score(model, df2.drop('Modal Price (Rs./Quintal)', axis=1), df2['Modal Price (Rs./Quintal)'], cv=k_folds)

mean_cv_score = cv_scores.mean()
std_cv_score = cv_scores.std()

print(f"CV Scores are: {cv_scores}")
print(f"Mean CV Score: {mean_cv_score}")
print(f"Std CV Score: {std_cv_score}")
```

```python

k_folds = KFold(n_splits=5, shuffle=True, random_state=42)

# Cross-validation
cv_scores = cross_val_score(model, x_train, y_train, cv=k_folds)

mean_cv_score = cv_scores.mean()
std_cv_score = cv_scores.std()

print(f"CV Scores are: {cv_scores}")
print(f"Mean CV Score: {mean_cv_score}")
print(f"Std CV Score: {std_cv_score}")
```

```python
# Cross-validation
cv_scores = cross_val_score(model, x_test, y_test, cv=k_folds)

mean_cv_score = cv_scores.mean()
std_cv_score = cv_scores.std()

print(f"CV Scores are: {cv_scores}")
print(f"Mean CV Score: {mean_cv_score}")
print(f"Std CV Score: {std_cv_score}")
```

Since there is no significant difference in the scores between training and validation data, so it can be conclude that there is no overfitting.
